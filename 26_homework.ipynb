```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt

import datetime as dt
```

```python
data = pd.read_excel('data/26.3 Когортный анализ.xlsx')
```

```python
data.head()
```

```python
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
```

```python
data['month_year'] = data.date.apply(get_month)
```

```python
data['cohort_month'] = data.groupby('user_id').month_year.transform("min")
```

```python
data.head()
```

```python
def get_date(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day 
```

```python
data_year, data_month, _ = get_date(data, 'date')
```

```python
cohort_year, cohort_month, _ = get_date(data, 'cohort_month')
```

```python
year_diff = data_year - cohort_year
month_diff = data_month - cohort_month
```

```python
data['cohort_index'] = year_diff * 12 + month_diff + 1
```

```python
data_id = data.pivot_table(index="cohort_month",
                         columns='cohort_index',
                         values='user_id',
                         aggfunc=len)
```

```python
first_column = data_id.iloc[:, 0]

relation_id = data_id.div(first_column, axis=0)
```

```python
data_sum = data.pivot_table(index='cohort_month',
                            columns='cohort_index',
                            values='price',
                            aggfunc=sum)
```

```python
first_column_sum = data_sum.iloc[:, 0]
relation_sum = data_sum.div(first_column_sum, axis=0)
```

```python
display(data_id)
display(data_sum)
```

```python
display(relation_id)
display(relation_sum)
```
