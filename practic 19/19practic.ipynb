```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Построить гистограмму
plt.figure(figsize=(8, 5)) # figsize — размерность фигуры, сначала указать ширину в дюймах, затем — высоту в дюймах
plt.hist(df.odometer, bins=15, color='green') # df.odometer — колонка, по которой строится гистограмма, bins — количество интервалов (по умолчанию 10), color — цвет графика
plt.xlabel('Количество пройденных миль') # легенда графика по горизонтали
plt.ylabel('Частота') # легенда графика по вертикали
plt.title('Распределение количества пройденных миль') # название графика
```

```python
stats = df.groupby(["year", "price_category"]).agg({"fuel": "first", "odometer": "mean"})
stats = stats.rename(columns={"fuel": "fuel_first", "odometer": "odometer_mean"})
stats
```

```python
#stats = df.groupby(["fuel"], as_index=False)[["id"]].count() # Одно и тоже
stats = df.groupby(["fuel"], as_index=False).agg({"id": "count"}).rename(columns={"id": "count"})
stats
```

```python
plt.bar(stats["fuel"], stats["count"], color=["orange", "red", "green"])
plt.xlabel("Тип топлива")
plt.ylabel("Количество")
plt.title("Распределение типов топлива")
```

```python
plt.figure(figsize=(12, 8))
plt.pie(stats["count"], labels=stats["fuel"], labeldistance=None, autopct="%1.0f%%")
plt.title("Распределение типов топлива")
plt.legend()
plt.show()
```

### Корреляция

```python

colors = {'low': "blue", "medium": "orange", "high": "black"}
fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors:
    color = colors[price_category]
    data = df[df["price_category"] == price_category]
    ax.hist(data["odometer"], bins=20, color=color, alpha=0.7)

plt.ylabel("Количество")
plt.xlabel("Пробег")
plt.title("Количество пройденных миль в зависимости от ценовой категории автомобиля")
plt.legend(colors, title="Ценовая категория")
plt.show()
```

```python
plt.figure()
plt.scatter(df["odometer"], df["year"])
plt.title("Зависимость пробега от года выпуска автомобиля")
plt.xlabel("Пробег")
plt.ylabel("Год выпуска")
plt.show()
```

```python
colors = {'low': "blue", "medium": "orange", "high": "black"}
fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors:
    color = colors[price_category]
    data = df[df["price_category"] == price_category]
    ax.scatter(data["odometer"], data['year'], color=color, label=price_category)

ax.legend(loc='upper right', title="Ценовая категория")
plt.ylabel("Пробег")
plt.xlabel("Год выпуска")
plt.show()
```
