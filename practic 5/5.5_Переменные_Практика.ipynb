1. Вы провели рекламную кампанию в соцсетях стоимостью 65 300 рублей. Она привлекла вам 228 новых клиентов. Создайте переменные для стоимости и количества новых клиентов. Используя созданные переменные, посчитайте значение CPA (cost per acquisition), которое рассчитывается как отношение стоимости кампании к количеству привлечённых клиентов. Выведите результат на экран.

```python
# Дополните код, чтобы создать переменные, и расчитайте CPA

cost = 65300
new_clients = 228
cpa = cost / new_clients
```

```python
# Допишите вывод итогового значения на экран
print(cpa)
```

2. В рекламе модного фитнес-центра вы увидели, что у них есть бассейн длиной 25 метров, шириной 15 метров и глубиной 2,5 метра. Вы стали размышлять, как долго обходить такой бассейн кругом, как много воды в него поместится и всё в таком духе. Создайте соответствующие переменные и посчитайте площадь поверхности бассейна, периметр поверхности и объём. Выведите все три параметра на экран, чтобы их легко можно было различить.

```python
# Дополните код
# Переменные из условия задачи:
length = 25
width = 15
depth = 2.5
```

```python
# Дополните код
# Рассчитайте площадь, периметр и объём (для объёма придумайте переменную самостоятельно)

area = length * width
perimeter = 2 * (length + width)
volume = length * width * depth

```

```python
# Дополните код
# Выведите на экран полученный результат для каждого из параметров

print("Площадь поверхности", area, "квадратных метров")
print("Периметр поверхности:", perimeter)
print("Объем:", volume)
```
